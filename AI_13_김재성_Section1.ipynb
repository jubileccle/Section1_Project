```python
import pandas as pd
# 저장한 파일 읽기
df1 = pd.read_csv('vgames.jsk.csv')
df1
```

```python
# 장르별 각 지역의 총 판매량
df1_genre = df1.groupby(df1["Genre"]).sum()
df1_genre
```

```python
df1_genre = df1_genre.drop(["Year"], axis = 'columns')
df1_genre
```

```python
# 년도별 각 지역의 총 판매량
df1_Year = df1.groupby(df1["Year"]).sum()
df1_Year
```

```python
import seaborn as sns

df = pd.DataFrame(df1_Year)
s = sns.barplot(x = "EU_Sales", data = df) # hue = "Product"
```
