```python
import pandas as pd

df = pd.read_csv('vgames2.csv')
df
```

```python
# 데이터의 결측치가 어디에 있는지 확인
df.isnull()
```

```python
# 결측값을 '0' 으로 대체
df.fillna(0)
```

```python
# 데이터 정보 확인
df.info()
```

```python
# 숫자 끝의 M 제거
def column_cleaning(col):
  temp_list = []

  for i in col:
    temp = i.replace("M","")
    temp_list.append(temp)
  return temp_list

df['NA_Sales'] = column_cleaning(df['NA_Sales'])
df['EU_Sales'] = column_cleaning(df['EU_Sales'])
df['JP_Sales'] = column_cleaning(df['JP_Sales'])
df['Other_Sales'] = column_cleaning(df['Other_Sales'])

df1 = df
df1
```

```python
# 숫자 끝의 K 제거 후 M 단위로 통일

def value_to_float(x):
    if 'K' in x:
        if len(x) > 1: # len(x) 리스트의 원소 개소, 리스트의 크기
            return float(x.replace('K', '')) * 0.001
        return x
    return x

df1['NA_Sales'] = df1['NA_Sales'].apply(value_to_float)
df1['EU_Sales'] = df1['EU_Sales'].apply(value_to_float)
df1['JP_Sales'] = df1['JP_Sales'].apply(value_to_float)
df1['Other_Sales'] = df1['Other_Sales'].apply(value_to_float)

df1
```

```python
# 각 컬럼 데이타 타입을 float 로 변환
df1['NA_Sales'] = df1['NA_Sales'].astype(float)
df1['EU_Sales'] = df1['EU_Sales'].astype(float)
df1['JP_Sales'] = df1['JP_Sales'].astype(float)
df1['Other_Sales'] = df1['Other_Sales'].astype(float)
df1.info()
```

```python
# 년도가 제대로 표시되지 않는 행 제거
df1 = df1[df1['Year'] > 1900]
df1
```

```python
# 판매량을 보기 편하도록 100을 곱해 소숫점을 제거
# 만단위(ten thousand)
df1['NA_Sales'] = df1['NA_Sales']*100
df1['EU_Sales'] = df1['EU_Sales']*100
df1['JP_Sales'] = df1['JP_Sales']*100
df1['Other_Sales'] = df1['Other_Sales']*100
df1
```

```python
# 판매량을 정수형 변환
df1['NA_Sales'] = df1['NA_Sales'].astype(int)
df1['EU_Sales'] = df1['EU_Sales'].astype(int)
df1['JP_Sales'] = df1['JP_Sales'].astype(int)
df1['Other_Sales'] = df1['Other_Sales'].astype(int)
df1
```

```python
# 함수 xx는 객체, args는 추가 변수 순위를 매길 컬럼과 상위 몇등까지 뽑을지 전달
def top_any(xx, args):
  top_row = xx.sort_values(by = args[0], ascending = False)[:args[1]]
  return top_row
```

```python
# 년도별 1위 판매량 장르
num_of_gen_by_year = df1.groupby("Year")[["Genre","Sum"]].apply(top_any, ["Sum", 1])
num_of_gen_by_year
```
